```python
import pandas as pd
```

```python
df = pd.read_csv("../data/aidev/java.csv")
df.shape
```

```python
codegptsensor = []
for row in df.itertuples():
    codegptsensor.append({
        "index": row.Index,
        "code": row.code,
        "contrast": "ABC",
        "label": row.label
    })
```

```python
contrastive_df = pd.DataFrame(codegptsensor)
```

```python
contrastive_df.shape
```

```python
contrastive_df.to_json("../data/aidev/java_codegptsensor.jsonl", orient="records", lines=True)
```

```python
python_df = pd.read_csv("../data/aidev/python.csv")
python_codegptsensor = []
for row in python_df.itertuples():
    python_codegptsensor.append({
        "index": row.Index,
        "code": row.code,
        "contrast": "ABC",
        "label": row.label
    })

python_codegptsensor_df = pd.DataFrame(python_codegptsensor)
python_codegptsensor_df.shape
python_codegptsensor_df.to_json("../data/aidev/python_codegptsensor.jsonl", orient="records", lines=True)
```
